2. What is a confusion table? What does it help us understand about a model's performance?

 A confusion table is a cross-tabulation of predicted and actual values. Since classification doesn't have a numerical residual to calculate, a confusion table provides a way to visualize success/failure through accuracy therefore measuring the model's performance

3. What does the SSE quantify about a particular model?

SSE quantifies the how good the fit of the model is by measuring the squared differences between the true values and predicted values.

4. What are overfitting and underfitting?

Underfitting occurs when the model is too simple to reliably explain the phenomenon you are interested in, and overfitting occurs when your model is too complex to reliably explain the phenomenon you are interested in. Underfitting can occur when a large k value is used causing little variance and a broad model smoothing out outliers. Overfitting can occur when a small k value is used causing high variance and an overconfident model sensitive to outliers.

5. Why does splitting the data into training and testing sets, and choosing $k$ by evaluating accuracy or SSE on the test set, improve model performance?

Splitting the data into training and testing sets and choosing k by evaluating accuracy or SSE on this test set allows the model to be trained on one portion of the data and tested on another portion is hasn't seen yet. By substituting uncertainty about future predictions for uncertainty about the hold-out data it allows us to test how well the model will perform with new data.

**Prediction**

strength - easy to interpret, direct/clear prediction

weakness - doesn't provide info on model's confidence, ambiguous on model's uncertainty

**Probability**

strength - quantifies model's uncertainty

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsRegressor

#1
raw_df = pd.read_csv('https://raw.githubusercontent.com/ashleynguyen04/knn/refs/heads/main/data/USA_cars_datasets.csv')
df = raw_df[['price', 'year', 'mileage']]

missing_values = df.isna().sum()
print(missing_values)

print(df.shape)
df.head()
```

```python
#2
def maxmin(z):
    w = (z-np.min(z))/(np.max(z)-np.min(z))
    return w

pd.to_numeric(df['year'],errors='coerce')
pd.to_numeric(df['mileage'],errors='coerce')

df.loc[:,'year'] = maxmin(df['year'])
df.loc[:,'mileage'] = maxmin(df['mileage'])
```

```python
#3
y = df['price'] # Create target variable vector
X = df.loc[:,['year','mileage']] # Create feature matrix
print(X.head())
print(X.describe())

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, # Feature and target variables
                                                    test_size=.2, # Split the sample 80 train/20 test
                                                    random_state=65) # For replication purposes
```

```python
#4

k_values = [3,10,25,50,100,300] # Different k values to try
SSE = [] # Store the SSE here

plt.figure(figsize=(12, 8))
for k in k_values:
    #kNN regression
    model = KNeighborsRegressor(n_neighbors=k)
    fitted_model = model.fit(X_train, y_train)
    y_hat = fitted_model.predict(X_test)
    SSE.append(np.sum( (y_test-y_hat)**2 ))

    #Scatterplot
    plt.subplot(2, 3, k_values.index(k) + 1)
    plt.scatter(y_test, y_hat, alpha=0.5)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()])
    plt.xlabel('Actual Price')
    plt.ylabel('Predicted Price')
    plt.title(f'Predicted Price vs Actual Price, k={k}')

plt.tight_layout()
plt.show()
```

```python
#5
SSE_min = np.min(SSE) # Find lowest recorded SSE
min_index = np.where(SSE==SSE_min) # Find the indices of SSE that equal the minimum
optiaml_k = min_index[0]+1 # Find the optimal value of k
print(SSE)
print(f'Optimal k: {optimal_k}')
```
